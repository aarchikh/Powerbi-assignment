# HEALTHCARE DATA ANALYSIS DASHBOARD

## Project Description

This project focuses on analyzing healthcare data to understand patterns related to patient conditions, treatment procedures, costs, satisfaction levels, and hospital readmissions.

The goal of this analysis is to extract meaningful insights that can help healthcare providers understand patient outcomes and improve treatment strategies.

Using Microsoft Power BI, an interactive dashboard was created to visualize key metrics such as treatment costs, patient satisfaction, age distribution, medical conditions, and procedure types. The dashboard allows users to filter and explore the data based on gender, age group, medical condition, and treatment procedure.

Overall, this project demonstrates how healthcare data can be transformed into actionable insights through data visualization and dashboard analytics.

 ## Dashboard Overview

The dashboard includes the following visualizations:

- **Sum of Cost by Condition** – Shows healthcare costs associated with different medical conditions.
- **Patient Satisfaction by Outcome and Procedure** – Analyzes satisfaction levels based on treatment outcomes.
- **Patient Readmission Indicator** – Displays the number of patients who were readmitted.
- **Age Distribution by Condition and Readmission** – Shows how age affects medical conditions and readmissions.
- **Cost by Procedure** – Compares costs of different medical procedures.
- **Patient Count and Satisfaction by Condition** – Analyzes the relationship between patient count and satisfaction levels.

## Key Insights

- Some medical conditions contribute significantly to healthcare costs.
- Patient satisfaction is generally higher when treatment outcomes are successful.
- Certain procedures are more expensive compared to others.
- Older age groups show higher chances of hospital readmission.

## Conclusion

This project highlights how healthcare data can be analyzed using visualization tools to better understand **patient outcomes, treatment effectiveness, and healthcare costs**. The dashboard provides a clear and interactive way to explore healthcare trends.
