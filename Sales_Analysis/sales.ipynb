# SALES DATA ANALYSIS DASHBOARD



## Project Description
This project analyzes sales data to understand product performance, regional sales distribution, and overall revenue trends. The analysis focuses on identifying top-selling products and comparing sales across different regions and product categories.

An interactive dashboard was created using Microsoft Power BI to visualize key sales metrics such as total sales by product, regional sales distribution, and overall revenue. Filters are included to explore sales by product category and region.

This project demonstrates how data visualization can help businesses monitor sales performance and identify important market trends.


## Dashboard Overview

The dashboard includes the following visualizations:

- **Total Sales by Product** – Displays the sales performance of different products.
- **Sales Distribution by Region** – Shows how sales are distributed across different regions.
- **Total Sales KPI** – Highlights the overall revenue generated.
- **Product Category Filter** – Allows users to filter sales by product categories.
- **Region Filter** – Enables analysis of sales across different geographic regions.
