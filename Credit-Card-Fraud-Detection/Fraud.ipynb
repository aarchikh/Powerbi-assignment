# CREDIT CARD FRAUD DETECTION 


## Project Description
This project analyzes financial transaction data to identify fraudulent transactions and understand fraud patterns. The analysis focuses on comparing fraud vs non-fraud transactions, monitoring transaction amounts, and identifying suspicious activity.

An interactive dashboard was created using Microsoft Power BI to visualize important metrics such as total transactions, total fraud cases, and transaction distribution. Filters are included to explore fraud and non-fraud records easily.

This project demonstrates how data visualization can help detect fraud patterns and support financial security analysis.



## Dashboard Overview

The dashboard includes the following visualizations:

- **Fraud vs Non-Fraud Transactions** – Compares the number of fraudulent and normal transactions.
- **Total Transactions** – Displays the total number of transactions in the dataset.
- **Total Fraud Transactions** – Shows the number of detected fraudulent transactions.
- **Transaction Table** – Displays transaction details such as fraud status, transaction amount, and transaction time.
- **Fraud Filter** – Allows users to filter transactions by fraud status.
