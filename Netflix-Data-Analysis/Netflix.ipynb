# NETFLIX DATA ANALYSIS DASHBOARD


## Project Description

This project analyzes the Netflix titles dataset to understand trends in the platform’s content library. The analysis focuses on key aspects such as the distribution of movies and TV shows, most popular genres, content ratings, monthly additions of titles, and the global distribution of Netflix content.

Using Tableau, an interactive dashboard was created to visualize these insights through charts, maps, and graphs. The project helps identify patterns in Netflix’s content growth, the countries contributing the most content, and the types of shows that dominate the platform.

Overall, this project demonstrates how data visualization can be used to extract meaningful insights from entertainment industry data and present them in a clear and interactive format.

## Dataset Information

The dataset used for this project contains information about Netflix movies and TV shows available on the platform.

Key features in the dataset include:
- Title
- Type (Movie or TV Show)
- Country
- Release Year
- Date Added
- Rating
- Genre
- Duration

## Tools Used

- Tableau – Dashboard creation
- Python / Jupyter Notebook – Project documentation
- Git & GitHub – Version control
- Kaggle – Dataset source

## Dashboard Overview

The Tableau dashboard provides visual insights into Netflix's content library.

The dashboard includes:
- Total Netflix titles
- Most popular genres
- Content added by month
- Growth of Netflix content over time
- Global distribution of Netflix content
- Top countries producing Netflix content
- Movies vs TV Shows distribution
- Content rating distribution

## Key Insights

- Netflix has over 8,800 titles in the dataset.
- Movies make up the majority of Netflix content.
- Drama and comedy are the most popular genres.
- The United States produces the highest amount of Netflix content.
- Netflix content increased significantly after 2015.

## Conclusion

This project demonstrates how data visualization can help identify trends in streaming platform content. The dashboard highlights Netflix’s global content distribution, popular genres, and rapid growth in recent years.
